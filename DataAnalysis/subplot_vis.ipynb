```python
# importing sub 1 box select data
sub_F_1_1_rot_for_x_DF = pd.read_csv("/Users/iman/CPSC_Courses/CPSC392ParlettPelleriti/Research/Data/Sub1/Freeform/session_F_RotForwardX_subjID_1_04-14-23_02-26-20.csv")
sub_F_1_1_rot_for_x_DF.drop(sub_F_1_1_rot_for_x_DF[(sub_F_1_1_rot_for_x_DF['trigger_pull_amount_left'] == 0) & (sub_F_1_1_rot_for_x_DF['trigger_pull_amount_right'] == 0)].index, inplace=True)
sub_F_1_1_rot_for_x_DF.insert(0, 'participant_num', 1)

# importing sub 2 box select data
sub_F_2_1_rot_for_x_DF = pd.read_csv("/Users/iman/CPSC_Courses/CPSC392ParlettPelleriti/Research/Data/Sub2/Freeform/session_F_RotForwardX_subjID_2_04-14-23_03-53-28.csv")
sub_F_2_1_rot_for_x_DF.drop(sub_F_2_1_rot_for_x_DF[(sub_F_2_1_rot_for_x_DF['trigger_pull_amount_left'] == 0) & (sub_F_2_1_rot_for_x_DF['trigger_pull_amount_right'] == 0)].index, inplace=True)
sub_F_2_1_rot_for_x_DF.insert(0, 'participant_num', 2)

# importing sub 3 box select data
sub_F_3_1_rot_for_x_DF = pd.read_csv("/Users/iman/CPSC_Courses/CPSC392ParlettPelleriti/Research/Data/Sub3/Freeform/session_F_RotForwardX_subjID_3_04-28-23_02-23-58.csv")
sub_F_3_1_rot_for_x_DF.drop(sub_F_3_1_rot_for_x_DF[(sub_F_3_1_rot_for_x_DF['trigger_pull_amount_left'] == 0) & (sub_F_3_1_rot_for_x_DF['trigger_pull_amount_right'] == 0)].index, inplace=True)
sub_F_3_1_rot_for_x_DF.insert(0, 'participant_num', 3)

# importing sub 4 box select data
sub_F_4_1_rot_for_x_DF = pd.read_csv("/Users/iman/CPSC_Courses/CPSC392ParlettPelleriti/Research/Data/Sub4/Freeform/session_F_RotForwardX_subjID_4_04-28-23_03-34-40.csv")
sub_F_4_1_rot_for_x_DF.drop(sub_F_4_1_rot_for_x_DF[(sub_F_4_1_rot_for_x_DF['trigger_pull_amount_left'] == 0) & (sub_F_4_1_rot_for_x_DF['trigger_pull_amount_right'] == 0)].index, inplace=True)
sub_F_4_1_rot_for_x_DF.insert(0, 'participant_num', 4)

# importing sub 5 box select data
sub_F_5_1_rot_for_x_DF = pd.read_csv("/Users/iman/CPSC_Courses/CPSC392ParlettPelleriti/Research/Data/Sub5/Freeform/session_F_RotForwardX_subjID_5_04-28-23_04-33-56.csv")
sub_F_5_1_rot_for_x_DF.drop(sub_F_5_1_rot_for_x_DF[(sub_F_5_1_rot_for_x_DF['trigger_pull_amount_left'] == 0) & (sub_F_5_1_rot_for_x_DF['trigger_pull_amount_right'] == 0)].index, inplace=True)
sub_F_5_1_rot_for_x_DF.insert(0, 'participant_num', 5)

print(sub_F_1_1_rot_for_x_DF.shape)
print(sub_F_2_1_rot_for_x_DF.shape)
print(sub_F_3_1_rot_for_x_DF.shape)
print(sub_F_4_1_rot_for_x_DF.shape)
print(sub_F_5_1_rot_for_x_DF.shape)

all_F_rot_for_x_DF = pd.concat([sub_F_1_1_rot_for_x_DF, sub_F_2_1_rot_for_x_DF, sub_F_3_1_rot_for_x_DF, sub_F_4_1_rot_for_x_DF, sub_F_5_1_rot_for_x_DF])
print(all_F_rot_for_x_DF.shape)
```

```python
grouped_by_participant = all_F_rot_for_x_DF.groupby('participant_num')

for participant_num, group in grouped_by_participant:
    fig = go.Figure()
    for trial_num, trial_group in group.groupby('gesture_counter'):
        fig.add_trace(go.Scatter3d(
            x=trial_group['r_controller_translation_x'],
            y=trial_group['r_controller_translation_z'],
            z=trial_group['r_controller_translation_y'],
            mode='markers',
            marker=dict(
                size=6,
                color=trial_group.index,  # set color based on occurrence (use index as a unique identifier)
                colorscale='Viridis',  # choose a colorscale
                opacity=0.8,
                colorbar=dict(
                    title='Right Occurrence',
                    tickvals=[trial_group.index[0], trial_group.index[-1]],
                    ticktext=['End', 'Start'],
                    len=0.6,
                    x=0.18,  # move the color bar to the left (adjust the value as needed)
                    y=0.8,
                    yanchor='top',
                )
            ),
            name=f'Trial {trial_num} - Right Controller',
            showlegend=False  # hide the legend entry for this trace
        ))
        
        fig.add_trace(go.Scatter3d(
            x=trial_group['l_controller_translation_x'],
            y=trial_group['l_controller_translation_z'],
            z=trial_group['l_controller_translation_y'],
            mode='markers',
            marker=dict(
                size=6,
                color=trial_group.index,  # set color based on occurrence (use index as a unique identifier)
                colorscale='sunset',  # choose a colorscale
                opacity=0.8,
                colorbar=dict(
                    title='Left Occurrence',
                    tickvals=[trial_group.index[0], trial_group.index[-1]],
                    ticktext=['End', 'Start'],
                    len=0.6,
                    x=0.001,  # move the color bar to the left (adjust the value as needed)
                    y=0.8,
                    yanchor='top',
                )
            ),
            name=f'Trial {trial_num} - Left Controller',
            showlegend=False  # hide the legend entry for this trace
        ))
    
    fig.update_layout(
        scene=dict(
            xaxis_title='X',
            yaxis_title='Z',
            zaxis_title='Y',
        ),
        title=f'Participant {participant_num}: Rotate Forward (Freeform)',
    )
    fig.show()
```
